### Testing Algorithms


The models I tested in this regression problem are a simple Linear Regression, Polyonimial Regression, Decision Tree Regressor, Random Forest Regressor, SVR, KRR, a simple neural network and PCA with linear regression. For metrics I used r2 score from sklearn in which 1 is the best possible score.

```python
diab_df = diab_df.drop(['s1'], axis=1)
X = diab_df.loc[:,diab_df.columns[0:9]]
Y = diab_df.loc[:,diab_df.columns[9]]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)
```

```python
def get_r2score(X_test, X_train, Y_test, Y_train, mod):
    
    Y_pred_test = mod.predict(X_test)
    Y_pred_train = mod.predict(X_train)
    return r2_score(Y_test, Y_pred_test), r2_score(Y_train, Y_pred_train)
```

#### 1) Linear Regression

```python
lr = LinearRegression().fit(X_train, Y_train)
lreg_cv= cross_validate(lr, X, Y, cv=5, scoring=('r2'),return_train_score=True, return_estimator=True)
test_r2_lr = np.mean(lreg_cv['test_score'])
train_r2_lr = np.mean(lreg_cv['train_score'])
```

#### 2) Polyonimial Regression

```python
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X_train)
X_test_ = poly_reg.fit_transform(X_test)
poly_reg_model = LinearRegression()

lreg_cv= cross_validate(poly_reg_model, X_poly, Y_train, cv=5, scoring=('r2'),return_train_score=True)
test_r2_pl = np.mean(lreg_cv['test_score'])
train_r2_pl = np.mean(lreg_cv['train_score'])
```

#### 3) Decisioin Tree Regressor

```python
dtr = DecisionTreeRegressor()
distributions = dict(max_depth=[3, 5, 10, 15], min_samples_split=[3, 5, 8, 15], max_features=[5, 7, 9])
clf = GridSearchCV(dtr, distributions)
search = clf.fit(X_train, Y_train)
print(search.best_params_)
test_r2_dtr, train_r2_dtr = get_r2score(X_test, X_train, Y_test, Y_train, clf)
```

#### 4) Random Forest Regressor

```python
rfr = RandomForestRegressor()
d = dict(n_estimators=[5, 10, 15], max_depth=[3, 5, 10, 15], min_samples_split=[ 3, 5, 8, 15], max_features=[5, 7, 9])
clf = GridSearchCV(rfr, d)
search = clf.fit(X_train, Y_train)
print(search.best_params_)
test_r2_rfr, train_r2_rfr = get_r2score(X_test, X_train, Y_test, Y_train, clf)
```

#### 5) SVR

```python
nusvr = NuSVR(C=1.0)
d = dict(nu=[0.2, 0.5, 0.7], kernel=['poly', 'rbf', 'sigmoid'])
clf = GridSearchCV(nusvr, d)
clf.fit(X_train, Y_train)
print(clf.best_params_)
test_r2_svr, train_r2_svr = get_r2score(X_test, X_train, Y_test, Y_train, clf)
```

#### 6) KRR

```python
import warnings 
warnings.filterwarnings('ignore')

krr = KernelRidge()
d = dict(alpha=[0.01, 0.5], kernel=['poly', 'rbf', 'sigmoid'])
clf = GridSearchCV(krr, d)
clf.fit(X_train, Y_train)
print(clf.best_params_)
test_r2_krr, train_r2_krr = get_r2score(X_test, X_train, Y_test, Y_train, clf)
```

#### 7) Neural Network

For neural network I used a simple multilayer perceptron. I tried with two or three layers and the best was (500, 200, 100), for activation functions I used relu and tanh and it chose relu, and for alpha the best choice seemed to be 0.0001. Also for optimizer I set 'adam'.

```python
mlp = MLPRegressor()
d = dict(hidden_layer_sizes=[(500, 200, 100), (500, 200)], activation = ['relu', 'tanh'], solver=['adam'], alpha=[0.0001, 0.001, 0.01])
clf = GridSearchCV(mlp, d) 
clf.fit(X_train, Y_train)
print(clf.best_params_)
test_r2_mlp, train_r2_mlp = get_r2score(X_test, X_train, Y_test, Y_train, clf)
```

#### 8) PCA

```python
diab_data =  load_diabetes()
diab_df = pd.DataFrame(data=diab_data.data, columns=diab_data.feature_names)
diab_df['target'] = pd.Series(diab_data.target)


X = diab_df.loc[:,diab_df.columns[0:10]]
Y = diab_df.loc[:,diab_df.columns[10]]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20)

best_test = 0
for n in [7, 8, 9]:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    X_train_pca_df = pd.DataFrame(data=X_train_pca, columns=np.arange(n))
    X_test_pca_df  = pd.DataFrame(data=X_test_pca, columns=np.arange(n))
    
    lr_ = LinearRegression().fit(X_train_pca_df, Y_train)
    test_r2_pca, train_r2_pca = get_r2score(X_test_pca_df, X_train_pca_df, Y_test, Y_train, lr_)
    
    if best_test < test_r2_lr:
        best_test_lr_pca = test_r2_pca
        train_r2_lr_pca = train_r2_pca
        best_n = n
print(best_n, best_test_lr_pca, train_r2_lr_pca)
```

```python
lr_ = LinearRegression()
pipe = Pipeline(steps=[('pca', pca), ('lr', lr_)])
param_dict = {"pca__n_components":[8, 9, 10]}
estimator = GridSearchCV(pipe, param_dict, verbose=2)
estimator.fit(X_train, Y_train)
print("The best parameters: {0}".format(estimator.best_params_))
pipe.set_params(**estimator.best_params_)
pipe.fit(X_train, Y_train)
test_r2_pca = pipe.score(X_test, Y_test)
train_r2_pca = pipe.score(X_train, Y_train)
```
